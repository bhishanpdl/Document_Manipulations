# Read ocr pdf using img2table (can also read native pdfs)

```python
import base64
from IPython.display import IFrame
from IPython.display import display

path_pdf = r"C:\Users\a126291\OneDrive - AmerisourceBergen(ABC)\data\pdf_files\screenshot_pages_eylea.pdf"
path_pdf = 'data/tables.pdf'
with open(path_pdf, 'rb') as fh:
    data_bytes = fh.read()

base64_pdf = base64.b64encode(data_bytes).decode('utf-8')
display(IFrame(src=f"data:application/pdf;base64,{base64_pdf}", width=800, height=800))
```

```python
from img2table.document import PDF
from img2table.ocr import TesseractOCR
tesseract_ocr = TesseractOCR(n_threads=1, lang="eng")

path_pdf = r"C:\Users\a126291\OneDrive - AmerisourceBergen(ABC)\data\module_img2table\data\tables.pdf"
pdf = PDF(src=path_pdf)

# Extract tables
dct_tables = pdf.extract_tables(ocr=tesseract_ocr,
                                      implicit_rows=False,
                                      borderless_tables=False,
                                      min_confidence=50)

dct_tables
```

```python
len(dct_tables)
```

```python
for page, tables in dct_tables.items():
    for idx, table in enumerate(tables):
        display(table.df)
```

```python
from IPython.display import display_html
for page, tables in dct_tables.items():
    for idx, table in enumerate(tables):
        display_html(table.html_repr(title=f"Page {page + 1} - Extracted table n°{idx + 1}"), raw=True)
```
